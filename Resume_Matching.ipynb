```python
# Importing necessary libraries
import os
import PyPDF2
import pandas as pd
from transformers import DistilBertTokenizer, DistilBertModel
from sklearn.metrics.pairwise import cosine_similarity
```

```python

# Function to extract PDF details
def extract_pdf_details(pdf_path):
    pdf_file = open(pdf_path, 'rb')
    pdf_reader = PyPDF2.PdfFileReader(pdf_file)
    num_pages = pdf_reader.numPages

    details = {
        'Category': '',
        'Skills': '',
        'Education': ''
    }

    for page_num in range(num_pages):  # Iterate through all pages
        page = pdf_reader.getPage(page_num)
        text = page.extractText()

        # Extract information based on expected keywords
        if 'Category' in text:
            details['Category'] = text.split('Category:', 1)[-1].strip()
        if 'Skills' in text:
            details['Skills'] = text.split('Skills:', 1)[-1].strip()
        if 'Education' in text:
            details['Education'] = text.split('Education:', 1)[-1].strip()

    pdf_file.close()
    return details

# Define the root directory where your folders are located
root_directory = '/content/data/data'

# List to store PDF details
all_cv_details = []

# List all subdirectories and their files
for root, dirs, files in os.walk(root_directory):
    for file in files:
        if file.endswith('.pdf'):
            pdf_path = os.path.join(root, file)
            details = extract_pdf_details(pdf_path)
            all_cv_details.append(details)

# Create a DataFrame from the extracted details
cv_df = pd.DataFrame(all_cv_details)

# Print the DataFrame
print(cv_df)
```

```python
import datasets
from datasets import load_dataset

dataset = load_dataset("jacob-hugging-face/job-descriptions")
job_descriptions = dataset['train']['job_description'][:15]
```

```python
# Step 3: Candidate-Job Matching
import torch
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
model = DistilBertModel.from_pretrained('distilbert-base-uncased')

def preprocess_text(text, max_seq_length=512):
    tokens = tokenizer.encode(text, add_special_tokens=True, max_length=max_seq_length, truncation=True)
    return tokens

def calculate_similarity(job_description, all_cv_details):
    job_description_tokens = preprocess_text(job_description)
    job_description_embedding = model(torch.tensor([job_description_tokens]))[0].mean(dim=1)

    similarities = []
    for cv in all_cv_details:
        cv_skills_tokens = preprocess_text(cv['Skills'])
        cv_skills_embedding = model(torch.tensor([cv_skills_tokens]))[0].mean(dim=1)

        # Convert tensors to NumPy arrays using detach()
        job_description_embedding_np = job_description_embedding.detach().numpy()
        cv_skills_embedding_np = cv_skills_embedding.detach().numpy()

        similarity = cosine_similarity(job_description_embedding_np, cv_skills_embedding_np)[0][0]
        similarities.append(similarity)

    return similarities
```

```python
top_ranked_cvs = []
for job_description in job_descriptions:
    similarities = calculate_similarity(job_description, all_cv_details)
    top_ranked_cvs.append(sorted(range(len(similarities)), key=lambda i: similarities[i], reverse=True)[:5])

for i, job_description in enumerate(job_descriptions):
    print(f"Job Description {i+1}: {job_description}")
    print("Top 5 CVs:")
    for cv_index in top_ranked_cvs[i]:
        print(all_cv_details[cv_index])
    print()
```
